```python
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
DATASET = "SonicNLP/medvqa_dataset"
data = load_dataset(DATASET)
```

# Train

```python
train_images = 3200
train_qa = 12792
val_images = 500
val_qa = 2000
test_images = 500
test_qa = 500 

# Plot train, val, test and add numbers of samples on top of bars
labels = ['Train', 'Validation', 'Test']
image_counts = [train_images, val_images, test_images]
qa_counts = [train_qa, val_qa, test_qa]
x = range(len(labels))
plt.bar(x, image_counts, width=0.4, label='Images', align='center')
plt.bar([i + 0.4 for i in x], qa_counts, width=0.4, label='Q&A', align='center')
plt.xticks([i + 0.2 for i in x], labels)
for i in range(len(labels)):
    plt.text(i, image_counts[i] + 50, str(image_counts[i]), ha='center')
    plt.text(i + 0.4, qa_counts[i] + 50, str(qa_counts[i]), ha='center')
plt.ylabel('Number of Samples')
plt.title('Dataset Sample Counts')
plt.legend()
plt.show()
```
